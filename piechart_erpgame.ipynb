```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
file_path = "https://github.com/MitaliP001/game/raw/main/sfinal_csv.csv"
df = pd.read_csv(file_path)

# Filter the data for SALES_ORGANIZATION 'OO'
filtered_df = df[df['SALES_ORGANIZATION'] == 'OO']

# Extract unique products from MATERIAL_DESCRIPTION column
selected_products = filtered_df['MATERIAL_DESCRIPTION'].unique()

# Pivot the data for better visualization using pivot_table
pivoted_df = pd.pivot_table(filtered_df, values='QUANTITY', index='SIM_ROUND', columns='MATERIAL_DESCRIPTION', aggfunc='sum')

# Plot the data as individual pie charts in one line
colors = plt.cm.Paired(range(len(selected_products)))
round_names = pivoted_df.index

fig, axs = plt.subplots(1, len(round_names), figsize=(5 * len(round_names), 5))

for i, round_name in enumerate(round_names):
    ax = axs[i]
    data = pivoted_df.loc[round_name]
    data.plot.pie(autopct='%1.1f%%', startangle=90, colors=colors, ax=ax)
    ax.set_title(f'Round {round_name}')
    ax.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular

# Add "Created by Mitali Purohit" text
plt.text(0.5, -0.1, 'Created by Mitali Purohit', transform=plt.gcf().transFigure, ha='center', color='gray')

plt.show()
```
